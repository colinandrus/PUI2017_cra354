```python
import numpy as np
import pylab as pl
```

```python
# Set random seed
np.random.seed(seed=39)
```

```python
# Generate size vector and mean

mean_for_all = 100
# Need to get enough near 10 to make show variation with small numbers
# But also need to be unique
low_nums = np.linspace(10, 65, num=50).astype(int)
high_nums = np.linspace(66, 2000, num=50).astype(int)
my_size = np.append(low_nums, high_nums)
print (my_size, mysize.shape)

```

```python
# Create dict to hold all distributions
dists={}
```

```python
# Normal Distribution
dists['normal'] = {} 

# Generate distributions
for n in my_size:
    dists['normal'][n] = np.random.normal(mean_for_all, size = n)

# Create key for means
dists['normal']['means'] = {}

# Plot
%pylab inline
ax_norm = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in dists['normal']:
    if not type(nn) == str:
        dists['normal']['means'][nn] = dists['normal'][nn].mean()
        ax_norm.plot(nn, dists['normal']['means'][nn], 'o')
        ax_norm.set_xlabel('sample size', fontsize=18)
        ax_norm.set_ylabel('sample mean', fontsize=18)
        ax_norm.set_title('Normal Distributions', fontsize=18)
        # Draw in line at mean
        ax_norm.plot([min(my_size), max(my_size)], [mean_for_all, mean_for_all], 'k')
```

# Discusion of Law of Large Numbers
For all of the distributions plotted above, there is a trend of increasing sample sizes leading to means closer to 100.  This is due to the law of large numbers which says the results of an an event will approach the expected value with more repetitions of the event.  The easiest one to visualize would be the binomial distribution.  Since we used a n of 200 and a p of 0.5 it was effectively simulating a fair coin flipping 200 times and counting the number of heads.  If we only do this one time, we most likely would not get 100 heads.  If we were to do this for thousands of times and then take the average of the number of heads from each set of 200 flips we would almost certainly have a number much closer to 100.
