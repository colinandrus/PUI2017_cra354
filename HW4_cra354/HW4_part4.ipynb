```python
# take may and january
df_jan = pd.read_csv('/gws/open/Student/citibike/201701-citibike-tripdata.csv.zip', compression='infer')
df_may = pd.read_csv('/gws/open/Student/citibike/201705-citibike-tripdata.csv.zip', compression='infer')
```

```python
# combine into one
frames = [df_jan, df_may]
df = pd.concat(frames)
```

```python
print(df.shape)
print(df_jan.shape)
print(df_may.shape)
```

```python
# looks good now make hour column
df['hour'] = pd.to_datetime(df['starttime']).dt.hour
```

```python
# only want rows with ok values for duration and hour
df = df[np.isfinite(df['hour'])]
df = df[np.isfinite(df['tripduration'])]
```

```python
df['hour'].value_counts()
```

```python
df.head()
```

```python
# some NaNs but seems alright to me
# I'll say between hours of 6pm and 4am is night
# hour >= 18 or hour < 4
# variable d_n day=0 night =1
def is_d_n(hour):
    if hour >= 18 or hour < 4:
        return(1)
    else:
        return(0)
df['d_n'] = df['hour'].apply(is_d_n)
```

```python
df['d_n'].value_counts()
#seems reasonable
```

```python
# now split by 'd_n' column
df_d = df[df['d_n'] == 0]
df_n = df[df['d_n'] == 1]
```
